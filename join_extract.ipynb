### Funções

```
def ler_dados(tabela: str) -> DataFrame:
    """
    Lê dados de uma tabela específica em um banco de dados PostgreSQL usando PySpark.

    Args:
        tabela (str): O nome da tabela a ser lida.

    Returns:
        DataFrame: Um DataFrame contendo os dados lidos da tabela.
    """
    
    print(tabela)
    driver = "org.postgresql.Driver"
    database_host = "psql-mock-database-cloud.postgres.database.azure.com"
    database_port = "5432"
    database_name = "ecom1689840238709wwavslrwfpjcxtfc"
    table = tabela
    user = "tpjwnzqctysgfpftkirsqpvy@psql-mock-database-cloud"
    password = "sownymlxwevaoowpwflbrcno"

    url = f"jdbc:postgresql://{database_host}:{database_port}/{database_name}"

    df = (spark.read
          .format("jdbc")
          .option("driver", driver)
          .option("url", url)
          .option("dbtable", table)
          .option("user", user)
          .option("password", password)
          .load())

    print(f'\nLinhas = {df.count()} \nColunas = {len(df.columns)}\n')
    df.printSchema()

    return df
```
